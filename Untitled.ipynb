```python
import lyricsgenius
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup

genius = lyricsgenius.Genius("lSUVIEFu57a_JUG5gUTMqPRgOUIBqNJJtSWWqaG1LnBc2FMi8qveGlBapwaTCRJe")
    
```

```python
artists = ['Kool Keith']

lyric_dict = {}
for art in artists:
    artist = genius.search_artist(art, max_songs=2,sort='title')
    for song in artist.songs:
        print(song.artist + ' ' + song.title)
        lyric_dict[song.artist + '_' + song.title] = song.lyrics
lyric_df = pd.DataFrame.from_dict(lyric_dict,orient='index').reset_index()
lyric_df.columns = ['song','lyrics']
cv = CountVectorizer()
df = cv.fit_transform(lyric_df['lyrics'])

count_vect_df = pd.DataFrame(df.todense(), columns=cv.get_feature_names())
lyric_df.join(count_vect_df)
#print(cv.vocabulary_ )
#df.todense()
#print(artist.songs(format = 'json'))
```
